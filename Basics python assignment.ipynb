```python
## checking the number is prime or not
```

```python
n=int(input("enter a number"))
a=0
for i in range(1,n+1):
    if(n%i==0):
        a=a+1
if(a==2):
    print("it is a prime number")
else:
    print("it is not a prime number")

```
