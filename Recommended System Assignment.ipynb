1. Can you explain the difference between user-based and item-based collaborative filtering?

```python
## User-Based CF -Finds similar users,Similarity between users,Hard to scale with many users,More users than items
## Item-Based CF -Finds similar items,Similarity between items,Easier to scale,More items than users
```
