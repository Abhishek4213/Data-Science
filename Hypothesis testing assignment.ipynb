```python
## hypothesis testing ##
```

```python
import numpy as np
import scipy.stats as stats
```

```python
#h0=model be accurate
#ha=may not be accurate
x_bar=3050
pop_mean=4000
n=25
std=125

#alpha 0.05,CI=0.95
```

```python
z_score=(x_bar-pop_mean)/(std/np.sqrt(n))
```

```python
z_score
```

```python
z_critical=stats.norm.ppf(1-0.025)
```

```python
z_critical
```

```python
margin=z_critical*(std/np.sqrt(n))
```

```python
margin
```

```python
mean_l=x_bar-margin
mean_u=x_bar+margin
```

```python
print(mean_l,',',mean_u)
```

```python
## z_critical > z_score 
#so the h0 be accepeted therefore the model be accurate
```

```python
###  Estimation and confidence interval
```

```python
n=15
data=[1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29]
```

```python
x_bar=(1.13+1.55+1.43+0.92+1.25+1.36+1.32+0.85+1.07+1.48+1.20+1.33+1.18+1.22+1.29)/15
```

```python
x_bar
```

```python
s=np.std(data,ddof=1)
s
```

```python
t_crit=stats.t.ppf(0.995,14)
t_crit
```

```python
margin_of_error=t_crit*(s/np.sqrt(n))
margin_of_error
```

```python
pop_mean_l=x_bar-margin_of_error
pop_mean_u=x_bar+margin_of_error
print(pop_mean_l,',',pop_mean_u)
```

```python
z_critical=stats.norm.ppf(0.995)
```

```python
z_critical
```

```python
margin = z_critical*(s/np.sqrt(n))
```

```python
margin
```

```python
mean_l=x_bar-margin
mean_u=x_bar+margin
```

```python
print(mean_l,',',mean_u)
```
